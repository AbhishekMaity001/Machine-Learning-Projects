## Handling the missing values

```python
# Finding the missing values Percentage for each and every column
np.round(df.isnull().mean(),2).sort_values(ascending=False)
```

```python
# Visualizing the missing values graphically

plt.figure(figsize=(15,10))
sns.heatmap(df.isnull(),cmap='viridis',yticklabels=False)
```

```python
df.head()
```

```python
# Since we will not be considering this as a time series data problem so dropping the City,Date,AQI_Bucket columns
df = df.drop(['City','Date','AQI_Bucket'],axis=1)
```

```python
df.head()
```

```python
# We can see that Xylene is totally missing for the Bengaluru so we will drop this column
df = df.drop(['Xylene'],axis=1)
```

```python
df['AQI'].isnull().sum() # our dependent variable is having 99 missing values so lets impute this with the median value
sns.histplot(df['AQI'],bins=50)
```

Since most of the variables looks like normally distributed lets impute the NaN values with the median of that particular column

```python
df.columns
```

```python
# There are many other ways of handling the missing values in the dataset according to the problem 
# To make it simple lets impute the NaN values with the median (we used median because of the outliers)

df['PM2.5'] = df['PM2.5'].fillna(df['PM2.5'].median())
df['PM10'] = df['PM10'].fillna(df['PM10'].median())
df['NO'] = df['NO'].fillna(df['NO'].median())
df['NO2'] = df['NO2'].fillna(df['NO2'].median())
df['NOx'] = df['NOx'].fillna(df['NOx'].median())
df['NH3'] = df['NH3'].fillna(df['NH3'].median())
df['CO'] = df['CO'].fillna(df['CO'].median())
df['SO2'] = df['SO2'].fillna(df['SO2'].median())
df['O3'] = df['O3'].fillna(df['O3'].median())
df['Benzene'] = df['Benzene'].fillna(df['Benzene'].median())
df['Toluene'] = df['Toluene'].fillna(df['Toluene'].median())
df['AQI'] = df['AQI'].fillna(df['AQI'].median())
```

```python
# Sanity check of the columns having null values or not
df.isnull().sum()
```

```python
sns.pairplot(df)
```

#### After Plotting the pair plot we can clearly see in the bottom row ....the relationship between all the independent variables and our dependent variable (AQI)

```python
sns.histplot(df['AQI'])
```

```python
# Dividing the dataset into independent (X) and dependent variable (y)
X = df.drop(['AQI'],axis=1)
y = df['AQI']
```

```python
# Plotting the correlation between different variables(sns.heatmap uses Pearson correlation coefficient)
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
```

```python
# Using the ExtraTreesRegressor for finding which all features are important

from sklearn.ensemble import ExtraTreesRegressor
extratreeregressor = ExtraTreesRegressor()
extratreeregressor.fit(X,y)
```

```python
extratreeregressor.feature_importances_
```

```python
feat_importances = pd.Series(extratreeregressor.feature_importances_,index = X.columns)
feat_importances.nlargest(12).plot(kind='barh')
plt.show()

#Since we have less number of features so we will not be dropping any features.
```

```python
# visualizing the distribution of the dependent variable
sns.distplot(y)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Note : i have tried many models on this dataset after preparing/cleaning the data so which ever model suits the best we will be going with that model

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X_train,y_train)
```

```python
# Linear Regression Model prediction
y_pred = linear_regression.predict(X_test)
print('R2 Score is  : {}'.format(linear_regression.score(X_train,y_train)))
print('R2 Score for test set is  : {}'.format(linear_regression.score(X_test,y_test)))
```

```python
# Evaluation Metrics

from sklearn import metrics 
print('MAE is : {}'.format(metrics.mean_absolute_error(y_test,y_pred)))
print('MSE is : {}'.format(metrics.mean_squared_error(y_test,y_pred)))
print('RMSE is : {}'.format(np.sqrt(metrics.mean_squared_error(y_test,y_pred))))
```

```python
# Visualizing the predicted value and the ground truth labels via scatter plot
sns.scatterplot(y_test,y_pred)
```

```python
import pickle

file = open('linear_regression_model.pkl','wb')
pickle.dump(linear_regression,file)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
random_model = RandomForestRegressor()
random_model.fit(X_train,y_train)
```

```python
# Random Forest Model
y_pred = random_model.predict(X_test)
print('R2 Score is  : {}'.format(random_model.score(X_train,y_train)))
print('R2 Score for test set is  : {}'.format(random_model.score(X_test,y_test)))
```

```python
# Evaluation metrics

from sklearn import metrics 
print('MAE is : {}'.format(metrics.mean_absolute_error(y_test,y_pred)))
print('MSE is : {}'.format(metrics.mean_squared_error(y_test,y_pred)))
print('RMSE is : {}'.format(np.sqrt(metrics.mean_squared_error(y_test,y_pred))))
```

For the Random Forest model it performed well with the train accuracy of 94% and test accuracy of 69% !! but we can do better with Hyperparameter tuning

```python
# Defining the parameters for the Random Forest Model
parameters = {
     'n_estimators' : [int(x) for x in range(100,1200,100)],
     'max_features' : ['auto','sqrt'],
     'max_depth' : [int(x) for x in range(2,11)],
     'min_samples_split' : [int(x) for x in range(1,6)],
     'min_samples_leaf' : [int(x) for x in range(1,10)]
 }
parameters    
```

```python
# Hyperpameter tuning for RandomForest Regressor using RandomizedSearchedCV
from sklearn.model_selection import RandomizedSearchCV
randomized = RandomizedSearchCV(random_model,param_distributions=parameters,n_iter=100,n_jobs=-1,cv=5
                               ,verbose=2,random_state=42)
randomized.fit(X_train,y_train)
```

```python
randomized.best_score_
```

```python
randomized.best_params_
```

```python
#Now lets predict thfitit test set with our new hyperparameterized model
y_pred = randomized.predict(X_test)
print('R2 Score is  : {}'.format(randomized.score(X_train,y_train)))
print('R2 Score for test set is  : {}'.format(randomized.score(X_test,y_test)))
```

```python
sns.distplot(y_test-y_pred)
```

```python
# Metrics

print('MAE : {}'.format(metrics.mean_absolute_error(y_test,y_pred)))
print('MSE : {}'.format(metrics.mean_squared_error(y_test,y_pred)))
print('RMSE : {}'.format(np.sqrt(metrics.mean_squared_error(y_test,y_pred))))
```

```python
sns.scatterplot(y_test,y_pred)
```
