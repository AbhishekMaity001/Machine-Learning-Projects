# Password Strength check 

```python
# importing all the necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the data...

df = pd.read_csv('data.csv',sep=',',error_bad_lines=False)
df.head()
```

```python
# shape of the data
df.shape
```

```python
df['strength'].value_counts() # checking the number of classes we are having.
```

```python
df[df['password'].isnull()] # checking for null values in dataset
```

```python
df.dropna(inplace=True) # Since we have only 1 Null value so Dropping the NaN value record
```

```python
df.isnull().sum() # Now we dont have any null values in our dataset.
```

```python
#converting the data frame into an array
password_tuple = np.array(df)
```

```python
password_tuple
```

```python
import random

#lets shuffle the passwords bcz these are in order....so for robustness lets shuffle
random.shuffle(password_tuple)
```

```python
password_tuple[0][1]
```

```python
# We can also shuffle these through pandas from iloc[] without converting it into an arrays
X = [label[0] for label in password_tuple] # Independent feature
y = [label[1] for label in password_tuple] # Dependent feature
```

```python
sns.countplot(y) # Checking the count of classes we have.
```

```python
X # Independent variable 
```

```python
#since my data is having more than 6 lakhs records ....so it is not possible to create vectors taking each unique words...
#Ram will be wasted...so we will do character by character

def word_divide_char(inputs):
    characters = []
    for i in inputs :
        characters.append(i)
    return characters
    
    
    
```

```python
word_divide_char([['Abhishek'],['Maity']])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# By default Tfidf takes words but here we are just overriding the tokenizer token and converting them into vectors
vectorizer = TfidfVectorizer(tokenizer=word_divide_char)
X = vectorizer.fit_transform(X)
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
X
```

```python
df.iloc[0,0]
```

```python
# code to see what is my vector for the 0th index password
feature_names = vectorizer.get_feature_names()

#get the tfidf for the first document
first_document_vector = X[0]

#print the vectors
dta = pd.DataFrame(first_document_vector.T.todense(),index=feature_names,columns=['TF-IDF'])
dta.sort_values(by=['TF-IDF'],ascending=False)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
log_classifier = LogisticRegression(penalty='l2',multi_class='ovr')
```

```python
log_classifier.fit(X_train,y_train)
```

```python
print(log_classifier.score(X_test,y_test))
```

```python
#Multinomial
clf = LogisticRegression(random_state=0,multi_class='multinomial',solver='newton-cg')
clf.fit(X_train,y_train)
print(clf.score(X_test,y_test))
```

```python
X_predict = np.array(['#$%abhishek@97'])
X_predict = vectorizer.transform(X_predict)
y_pred = log_classifier.predict(X_predict)
print(y_pred)
```
