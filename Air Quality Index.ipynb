```python
df['AQI'].isnull().sum() # our dependent variable is having 99 missing values so lets impute this with the median value
sns.histplot(df['AQI'],bins=50)
```

Since most of the variables looks like normally distributed lets impute the NaN values with the median of that particular column

```python
df.columns
```

```python
# To make it simple we only imputed the NaN values with the median

df['PM2.5'] = df['PM2.5'].fillna(df['PM2.5'].median())
df['PM10'] = df['PM10'].fillna(df['PM10'].median())
df['NO'] = df['NO'].fillna(df['NO'].median())
df['NO2'] = df['NO2'].fillna(df['NO2'].median())
df['NOx'] = df['NOx'].fillna(df['NOx'].median())
df['NH3'] = df['NH3'].fillna(df['NH3'].median())
df['CO'] = df['CO'].fillna(df['CO'].median())
df['SO2'] = df['SO2'].fillna(df['SO2'].median())
df['O3'] = df['O3'].fillna(df['O3'].median())
df['Benzene'] = df['Benzene'].fillna(df['Benzene'].median())
df['Toluene'] = df['Toluene'].fillna(df['Toluene'].median())
df['AQI'] = df['AQI'].fillna(df['AQI'].median())
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
random_model = RandomForestRegressor()
random_model.fit(X_train,y_train)
```

```python
# Random Forest Model
y_pred = random_model.predict(X_test)
print('R2 Score is  : {}'.format(random_model.score(X_train,y_train)))
print('R2 Score for test set is  : {}'.format(random_model.score(X_test,y_test)))
```

```python
# Linear Regression Model
y_pred = linear_regression.predict(X_test)
print('R2 Score is  : {}'.format(linear_regression.score(X_train,y_train)))
print('R2 Score for test set is  : {}'.format(linear_regression.score(X_test,y_test)))
```

# Evaluation Metrics

```python
from sklearn import metrics 
print('MAE is : {}'.format(metrics.mean_absolute_error(y_test,y_pred)))
print('MSE is : {}'.format(metrics.mean_squared_error(y_test,y_pred)))
print('RMSE is : {}'.format(np.sqrt(metrics.mean_squared_error(y_test,y_pred))))
```
